#  5. 列表(list)

1.列表本质: 有序的“万能储物筐”，用[]定义，元素间用逗号分隔，支持任意数据类型。  

2.核心操作:  
- 访问:索引(取单个)、切片(取部分)，规则和字符串一致;  
- 修改:直接通过列表[索引]= 新值修改元素;  
- 嵌套:列表里装列表，通过“外层索引+内层索引”访问嵌套元素  

3.常用函数:  

- 添加:append(加1个)、extend()(加多个)、insert()(指定位置加);  
- 查找:index()(找索引),count()(统计次数);  
- 删除:remove()(删第一个匹配值);  
- 排序: sort()(升/降序)、reverse() (反转)。    

4.关键提醒: 列表是“可修改”的，操作后直接改变原列表;函数操作时注意不存在的元素会报错。  

````python
list=["a","b","c"]
print(list)
list=[1,2,3]
print(list)
list.extend([4,5,6])
print(list)
````

#  7. 字典 （键值操作） 
注：键必须唯一，不可重复！ 

1. **字典本质**：有序的“键值对容器”，用 `{key: value}` 定义，键唯一且不可变，值可任意。

2. **核心操作**：
   - **查找**：优先用 `get(key, default)`，安全不报错；`dict[key]` 找不到会报错；
   - **增改**：`字典名[键] = 值`，键存在则改，不存在则加；
   - **删除**：`pop(key)` 删指定键值对，`clear()` 清空字典；
   - **遍历**：`keys()` 取所有键、`values()` 取所有值、`items()` 取所有键值对。

3. **关键优势**：通过键快速查找值，不用遍历，比列表查询效率高，适合存储关联数据。

4. **实用技巧**：存储用户信息、配置参数、映射关系时，优先用字典，逻辑更清晰。


````python
dfc={"name":"张三","age":18}
print(dfc)
````

# 11. 异常处理

## 1. 新手常见异常

| 异常类型            | 通俗原因                         | 示例场景 |
|---------------------|----------------------------------|----------|
| `NameError`         | 用了没定义的变量                 | `print(num)`（num 没赋值） |
| `SyntaxError`       | 代码语法错（不符合 Python 规则） | `if 1 > 2`（少了冒号） |
| `IndexError`        | 索引 / 下标超出范围              | `li = [1,2,3]; print(li[3])`（最大索引是 2） |
| `ZeroDivisionError` | 除数为 0                         | `print(5/0)` |
| `KeyError`          | 字典里没有这个键                 | `info={"name":"小明"}; print(info["age"])` |
| `IOError`           | 文件操作失败（如文件不存在）     | 尝试打开一个不存在的 `.txt` 文件 |
| `AttributeError`    | 对象没有这个属性                 | 字符串调用列表的 `append()` 方法 |
| `ValueError`        | 值不对（类型对但内容无效）       | `int("abc")`（字符串不能转整数） |
| `ImportError`        | 导入模块失败(路径错或名字错)       | `import pandas123`模块名不存在 |
| `IndentationError`        | IndentationError       | `if`语句后没缩进 |  \
 
## 2. 格式
try:  
except (value) as e:  
else:  
finally:  



````python
try:
    print(10/0)
except ZeroDivisionError as e:
    print("发生了除零错误：", e)
````

````python
try:
    print("文件操作")
except Exception as e:
    print("发生了错误：", e)
finally:
    print("关闭文件")
````

## 3. 异常处理的核心原则

1. **优先捕获具体异常**：  
   比如明确会触发 `ValueError`，就只捕获它，别用 `Exception` 盲目捕获所有异常（可能掩盖未知错误）。

2. **不要滥用异常处理**：  
   语法错误（比如少冒号、缩进错）要提前修正，不能靠 `try-except` 掩盖。

3. **处理逻辑要友好**：  
   给用户清晰的提示（比如“请输入纯数字”），而不是只打印“出错了”。

4. **`finally` 用于释放资源**：  
   比如关闭文件、断开网络连接，确保资源不会泄露。

# 12. 自定义异常

## 自定义异常的实现步骤（两步搞定）

步骤如下：

- **步骤 1：创建异常对象**  
  语法：`异常类型(异常具体描述信息)`

- **步骤 2：用 `raise` 语句抛出异常对象**  
  语法：`raise 异常对象`

````python
# 银行取款场景
balance = 1000  # 账户余额
withdraw_amount = 1500  # 取款金额
is_frozen = False  # 账户是否冻结
def withdraw(amount):
    global balance

    if is_frozen:
        raise Exception("账户被冻结，无法取款")
    if amount > balance:
        raise Exception("余额不足，无法取款")
    balance -= amount
    return balance
try:
    print(withdraw(withdraw_amount))
except Exception as e:
    print("发生了错误：", e)
finally:
    print(balance)
````

# 13. 模块与导入

## 1. 关键技巧：让模块测试代码不影响导入

写模块时，我们会加测试代码（比如调试 `add` 函数是否正常），  
但导入模块时，测试代码会自动执行，很麻烦。

可以用下面的写法解决：

```python
if __name__ == "__main__":
    # 测试代码
    pass  

- 模块直接运行时，`__name__` 的值是 `"__main__"`；
- 模块被导入时，`__name__` 的值是模块名（比如 `my_tools`）；
- 所以 `if __name__ == "__main__":` 里的代码，只有直接运行模块时才会执行。

## 13.1 导入模块的注意事项

1. **导入语句建议写在文件开头**：  
   方便阅读和组织代码，符合 Python 编程习惯。

2. **模块只导入一次**：  
   同一个模块在程序中只需导入一次，首次导入后会加载到内存，  
   后续导入不会重复加载（节省资源，避免冲突）。

3. **避免命名冲突**：
   - 两个模块有同名功能时，用 `模块名.功能名` 区分，或用 `as` 设别名；
   - 尽量不用 `from 模块名 import *`，容易覆盖当前文件的变量 / 函数。

## 13.2 核心总结

1. **模块本质**：  
   一个 `.py` 文件，用于封装变量、函数、类，实现代码复用和模块化管理。

2. **模块分类**：  
   - 内置模块（直接用）  
   - 第三方模块（`pip` 安装后用）  
   - 自定义模块（自己写）

3. **3 种导入方式**：
   - `import 模块名 as 别名`：导入所有功能，用“别名.功能名”调用，无冲突；
   - `from 模块名 import 功能名`：导入指定功能，直接调用，更简洁；
   - `from 模块名 import *`：导入所有功能，直接调用，但易冲突，慎用。

4. **关键技巧**：  
   用 `if __name__ == "__main__":` 写模块测试代码，不影响导入。

5. **避坑点**：  
   模块名不能和内置模块、第三方模块重名，否则功能可能失效。

# 14. 包  
#### 包是包含多个相关模块(.py文件)的文件夹


## 14.1 包的核心标志：`__init__.py` 文件

一个文件夹要被 Python 识别为包，必须包含 `__init__.py` 文件  
（Python 3.3+ 可省略，但建议保留，明确告诉 Python “这是个包”）。

### 这个文件的 3 个核心作用：

1. **标识目录为包**：  
   告诉 Python 这个文件夹不是普通文件夹，而是一个包。

2. **控制导入范围**：  
   通过 `__all__` 变量，指定批量导入时要包含的模块。

3. **执行初始化操作**：  
   导入包时，会自动执行这个文件里的代码  
   （比如初始化变量、导入必要模块）。

## 14.2 方式
`from 包名 import *` —— 批量导入（需配置 `__all__`）  
#### 指定批量导入时要包含的模块（写模块名，不带 .py）
__all__ = ["calc", "file_tool"]  
## 14.3 语法：
```python
from 包名 import *  

- 注意：默认情况下，这种方式只会导入 __init__.py 里直接定义的内容，
- 不会导入包内的模块，需要用 __all__ 变量指定要导入的模块。

## 14.4 核心总结

1. **包的本质**：  
   包含 `__init__.py` 文件的文件夹，用于组织多个相关模块，避免命名冲突。

2. **包的创建**：  
   新建包 → 放入相关模块。

3. **3 种常用导入方式**：
   - `import 包名.模块名`：层级清晰，无冲突，适合复杂项目；
   - `from 包名 import 模块名`：简化书写，不用带包名；
   - `from 包名 import *`：批量导入，需要在 `__init__.py` 中配置 `__all__`。

4. **关键文件**：  
   `__init__.py` 用于标识包、控制导入范围、执行初始化。

5. **实用场景**：  
   项目代码较多时，用包按功能分类（比如“用户模块包”“订单模块包”），  
   让代码结构更清晰，便于维护。
