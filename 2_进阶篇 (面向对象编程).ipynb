# 1. 类

## 1.1 构造函数 `__init__` : 创建对象时自动初始化 

### 1. 核心作用

创建对象时，Python 会自动调用 `__init__` 方法，  
给对象的实例属性设置初始值，无需手动赋值。

---

### 2. 语法

```python
class 类名:
    def __init__(self, 参数1, 参数2, ...):
        # 给实例属性赋值（self.属性名 = 参数）
        self.属性1 = 参数1
        self.属性2 = 参数2

    # 其他方法...

````python
class Car:
    def __init__(self, color, brand, year):
        self.color = color
        self.brand = brand
        self.year = year
    
car1 = Car("red", "Toyota", 2020)
print(car1.color)  # 输出: red
car2 = Car("blue", "Honda", 2019)   
print(car2.brand)  # 输出: Honda
````

### 3. 核心总结

1. **面向对象核心**：  
   以“对象”为中心，注重“谁来做”，适合复杂、可复用的任务。

2. **类和对象**：
   - 类：抽象的“设计图”，包含属性（特征）和方法（行为）；
   - 对象：类的“成品”，真实存在，能调用方法、访问属性；
   - 关系：先定义类，再创建对象。

3. **类的组成**：
   - 方法：类里的函数（比如 `start()`），实例方法第一个参数必须是 `self`，由对象调用；
   - 属性：类属性（所有对象共享）和实例属性（每个对象独有）。

4. **构造函数 `__init__`**：  
   创建对象时自动调用，用于初始化实例属性，简化赋值流程。

5. **关键语法**：
   - 定义类：`class 类名: ...`（大驼峰命名）；
   - 创建对象：`对象名 = 类名(参数)`（参数传给 `__init__`）；
   - 调用方法 / 访问属性：`对象名.方法名()`、`对象名.属性名`。

# 3. 继承（类之间的父子关系）


继承是类之间的“父子关系”：

- **父类（基类）**：被继承的类（比如“动物类”）；
- **子类（派生类）**：继承父类的类（比如“狗类”“猫类”）；


### 📌 核心：

子类可以：
- 直接使用父类的**非私有属性和方法**；
- 添加自己的**新属性 / 方法**；
- **重写（覆盖）父类的方法**。

````python
class Animal:
    def __init__(self, species, age):
        self.species = species
        self.age = age

class Dog(Animal):
    def __init__(self, species, age, breed):
        super().__init__(species, age)          # 调用父类的构造方法
        self.breed = breed
````

## 3.1 覆盖父类方法（完全替换）

当子类的方法实现和父类完全不同时，  
可以直接重写方法，调用时执行子类的实现。

---

### 示例：动物 → 猫（重写“发出声音”方法）

````python
class Animal:
    def speak(self):
        print("动物在叫")

class Cat(Animal):
    def speak(self):
        print("猫在喵喵叫")

c = Cat()
c.speak()  # 输出：猫在喵喵叫
````

## 3.2 扩展父类方法（保留 + 新增）

当子类需要**保留父类方法的功能**，再添加新逻辑时，  
使用 `super()` 调用父类方法。

---

### 语法：
```python
super().父类方法名()  # 推荐写法，不依赖父类名

````python
class Animal:
    def speak(self):
        print("动物在叫")

class Dog(Animal):
    def speak(self):
        super().speak()  # 调用父类的方法
        print("狗在汪汪叫")
````

## 3.3 多继承的坑：同名方法冲突

如果多个父类有同名方法，子类调用时会遵循“继承顺序”（MRO），  
即按子类定义时父类的排列顺序，优先执行左侧父类的方法。

---

### 示例：手机 + 相机 → 拍照手机（同名 `take_photo` 方法）


````python
class Phone:
    def take_photo(self):
        print("用手机拍照")

class Camera:
    def take_photo(self):
        print("用相机拍照")

class SmartPhone(Phone, Camera):
    pass

sp = SmartPhone()
sp.take_photo()  # 输出：用手机拍照（优先左边 Phone）
Camera.take_photo(sp)  # 输出：用相机拍照（调用 Camera 的方法）
````

## 3.4 多继承的弊端

虽然多继承能整合多个功能，但不推荐使用：

1. **冲突风险高**：  
   同名方法 / 属性容易混淆，依赖继承顺序，容易出错。

2. **代码复杂度高**：  
   父类过多时，层级关系混乱，不好理解和维护。

3. **耦合性强**：  
   子类和多个父类绑定，父类修改会影响子类。

---

### ✅ 替代方案

使用 **单继承 + 组合**：

👉 在子类中创建其他类的实例来复用功能，而不是继承多个父类。

## 3.5 核心总结

1. **继承本质**：  
   子类复用父类的非私有属性和方法，减少重复代码。

2. **单继承**：  
   子类只继承一个父类，语法：`class 子类(父类):`，支持传递性。

3. **方法重写**：
   - **覆盖**：子类定义同名方法，完全替换父类功能；
   - **扩展**：用 `super().方法名()` 调用父类方法，再新增逻辑。

4. **多继承**：  
   子类继承多个父类，语法：`class 子类(父类1, 父类2):`，  
   存在同名方法冲突风险，尽量不用。

5. **关键规则**：
   - **继承顺序（MRO）**：
     - 单继承：子类 → 父类 → 更上层父类；
     - 多继承：子类 → 左侧父类 → 右侧父类 → `object`；
   - **`super()`**：用于扩展父类方法，不依赖父类名，推荐使用。

# 4. 多态 -- 核心思想 "同一接口，多种实现"


   ## 4.1 什么是多态  
   不同类的对象，调用通明方法时，会执行各自类的实现，得到不同的结果。 
   ## 4.2 多态的核心价值  
- **增强代码可读性**：  
  让代码逻辑更清晰，更符合人类的思维习惯；

- **统一接口**：  
  无论对象的具体类型是什么，都可以通过同名方法调用  
  （如动物的 `make_sound()`、支付的 `pay()`）；

- **隐藏差异**：  
  调用者不需要关心对象的具体类型，只需调用统一接口，降低使用难度；

- **便于扩展**：  
  新增对象类型时，不用修改原有代码，直接实现新类 + 同名方法即可。  


## 4.3 方式 1：鸭子类型（Python 特有，最常用）

**核心逻辑：**  
“鸭子类型”是 Python 的特色思想：

> 不管你是什么类的对象，只要你有某个方法，我就可以用这个方法调用你，不用关心你到底是谁。

就像一句经典比喻：  
👉 “走路像鸭子、叫起来像鸭子，那就把你当鸭子。”

---

**示例：电商支付系统（微信 + 支付宝 + 银行卡）**

````python
class Wechatpay:
    def pay(self):
        print("使用微信支付")

class Alipay:
    def pay(self):
        print("使用支付宝支付")

class bankpay:
    def pay(self):
        print("使用银行卡支付")

# 统一接口
def process_payment(payment_method):
    payment_method.pay()

process_payment(Wechatpay())  # 输出：使用微信支付
process_payment(Alipay())    # 输出：使用支付宝支付 
process_payment(bankpay())   # 输出：使用银行卡支付
````

## 4.4 方式 2：继承 + 方法重写（适合有层级关系的场景）

**核心逻辑：**  
子类继承父类后，重写父类的同名方法，不同子类有不同实现。  
调用时，传入不同子类对象，执行对应重写的方法（体现多态）。

---

**示例：动物叫声系统（动物 → 狗 / 猫 / 鸡）**

````python
class Animal:
    def speak(self):
        print("动物在叫")

class Dog(Animal):
    def speak(self):
        print("狗在汪汪叫")

class Cat(Animal):
    def speak(self):
        print("猫在喵喵叫")

def make_animal_speak(animal):
    animal.speak()

make_animal_speak(Dog())  # 输出：狗在汪汪叫
make_animal_speak(Cat())  # 输出：猫在喵喵叫    
````

## 5. 核心总结

1. **多态本质**：  
   同一方法名，不同对象有不同实现，调用时产生不同结果。

2. **两种实现方式**：
   - **鸭子类型（推荐）**：  
     不用继承，只要对象有同名方法，就能调用；
   - **继承 + 方法重写**：  
     子类继承父类，重写同名方法，适合有层级关系的场景。

3. **核心接口**：  
   多态依赖“统一方法名”作为接口，调用者通过接口操作对象，不用关心对象类型。

4. **关键优势**：  
   统一接口、便于扩展、提高代码可读性。

5. **Python 特色**：  
   鸭子类型让多态更灵活，不用像其他语言（如 Java）那样依赖抽象类或接口，  
   对新手更友好、上手更快。

# 5. 静态方法


引言: 之前学的实例方法(带self参数)需要创建对象才能调用，而静态方法和类方法是类的两种特殊!  
- 方法：不用创建对象也能调用，还能分别处理“独立功能”和“类相关功能”。  
学会这两种方法，能避免“为了调用一个简单功能而被迫创建对象”的麻烦，让类的功能划分更清晰，代码更简洁。

## 5.1 什么是静态方法？

静态方法是定义在类里的“独立工具函数”，和类本身、实例（对象）都没有关系 ——  
不用访问类属性，也不用访问实例属性，只靠传入的参数就能完成功能。


📌 类比理解：

就像工具箱里的扳手：  
扳手是工具箱的一部分，但使用时不需要关心工具箱本身的状态  
（比如工具箱的颜色、大小），只要有要拧的螺丝（参数）就能直接用。

---

## 5.2 定义与调用

### 5.2.1 定义格式

```python
class 类名:
    @staticmethod
    def 方法名(参数1, 参数2, ...):
        # 不使用 self / cls
        pass

### 5.2.2 调用方式

- **推荐：通过类调用**
  
  直接使用 `类名.方法名()` 调用，不需要创建对象。

- **可选：通过对象调用**
  
  也可以用 `对象.方法名()` 调用（不推荐，因为静态方法和实例无关）。


````python
import math
class Geometrytools:
    @staticmethod
    def calculate_circle_area(radius):
        return math.pi * radius ** 2

print(Geometrytools.calculate_circle_area(5))  # 输出：78.53981633974483
````

## 5.2.3 应用场景

- **实现通用工具功能**：  
  比如格式验证、数学计算、数据转换等；

- **无需访问类/实例属性**：  
  方法只依赖参数即可完成任务；

- **优化代码结构**：  
  将独立函数归类到某个类中，让代码更清晰  
  （例如把所有几何计算函数放到 `GeometryTool` 类中）。

# 6. 类方法：操作类属性的类级方法


## 6.1 什么是类方法？

类方法是和“类本身”绑定的方法，专门用来操作类属性  
（所有对象共享的属性），而不是实例属性。

它接收的第一个参数是 `cls`（代表类本身），  
通过 `cls` 可以访问或修改类属性。

---

📌 类比理解：

就像班级的“班级管理函数”：  
负责处理班级的公共信息（例如全班人数、班级名称），  
而不是某个学生的个人信息（例如姓名、成绩）。

### 6.1.1 定义格式  
```python
class 类名  
    @classmethod            # 标志 不能少
    def 方法名(cls,参数列表):   # 第一个参数cls
        方法体


### 6.2.2 调用方式

- **推荐：通过类调用**

  直接使用 `类名.方法名()` 调用。

- **可选：通过对象调用**

  也可以使用 `对象.方法名()` 调用  
  （因为 `cls` 依然代表类本身，和具体实例无关）。

---

## 示例：人类（操作类属性）

````python
class Person:
    average_age = 70  # 类属性，表示平均年龄

    @classmethod
    def get_average_age(cls):
        return cls.average_age

    @classmethod
    def set_average_age(cls, age):
        cls.average_age = age

print(Person.get_average_age())  # 输出：70
Person.set_average_age(75)
print(Person.get_average_age())  # 输出：75
````

## 6.2.3 应用场景

- **访问 / 修改类属性**：  
  比如获取或更新全局共享数据（如全班人数、产品默认价格）；

- **工厂方法（推荐重点）**：  
  封装对象创建逻辑，提供多种创建实例的方式  
  （例如根据不同参数创建不同对象）；

- **类级别工具函数**：  
  需要用到类本身信息的功能  
  （例如从配置文件初始化类属性）。

| 对比维度  | 静态方法（@staticmethod）        | 类方法（@classmethod）          | 实例方法                |
| ----- | -------------------------- | -------------------------- | ------------------- |
| 第一个参数 | ❌ 无（不需要 `self / cls`）      | ✅ `cls`（代表类）               | ✅ `self`（代表实例）      |
| 装饰器   | `@staticmethod`            | `@classmethod`             | ❌ 不需要               |
| 访问属性  | ⚠️ 只能访问类属性（一般不推荐）          | ✅ 访问类属性（通过 `cls`）          | ✅ 访问实例属性（通过 `self`） |
| 调用方式  | `类名.方法()`（推荐）<br>`对象.方法()` | `类名.方法()`（推荐）<br>`对象.方法()` | `对象.方法()`（必须创建对象）   |
| 核心用途  | 工具函数（计算、验证等）               | 操作类属性、工厂方法                 | 操作实例数据              |
