# 1. 正则表达式使用

**引言：**  
正则表达式是 Python 里的“字符串匹配神器”——能快速查找、验证、提取符合特定规则的字符串，比如验证手机号格式、提取文本中的邮箱、过滤无效字符。虽然它的语法看起来复杂，但核心是“用规则描述模式”。学会使用正则表达式，能摆脱繁琐的字符逐个判断，轻松搞定各种字符串处理难题。

## 1. 正则表达式基础

### 1.1 什么是正则表达式？

正则表达式（简称“正则”）是一套描述字符串模式的规则，就像“字符串过滤器”——告诉程序“你要找的字符串长什么样”，程序就能按规则匹配出符合要求的内容。

**比如：**

- 规则 `\d{11}`：匹配 11 位数字（可用于验证手机号）；
- 规则 `\w+@\w+\.\w+`：匹配邮箱格式（比如 `abc123@qq.com`）。

### 1.2 核心用途

- **验证**：判断字符串是否符合规则（比如手机号、邮箱、身份证号格式是否正确）；
- **查找**：从文本中找出所有符合规则的字符串（比如从新闻中提取所有日期）；
- **提取**：从匹配结果中提取关键信息（比如从邮箱中提取用户名）；
- **替换**：替换文本中符合规则的内容（比如过滤文本中的特殊字符）。

### 1.3 Python 中使用正则的前提

Python 内置 `re` 模块（不用安装），所有正则操作都通过这个模块实现，第一步必须导入：

```python
import re

### 1.4 正则入门：3步完成匹配  
使用正则表达式的核心流程:定义规则一执行匹配一提取结果，以最基础的`re.match()`为例(从字符串开头开始匹配)  
示例：匹配h开头的字符串

````python
import re
pattern = "h\w+"        # \w表示匹配字母、数字或下划线，+表示匹配前一个字符一次或多次
string1 = "hello world"
string2 = "welcome to python"
result1 = re.match(pattern, string1)
print(result1.group())
````

### 1.5 常用规则  
### 常用规则符号

| 规则符号 | 含义 | 示例 | 匹配结果 |
|----------|------|------|----------|
| `.` | 匹配除换行符（`\n`）外的任意 1 个字符 | `re.match(".", "a")` | 匹配成功（"a"） |
| `[abc]` | 匹配 "a"、"b"、"c" 中的任意 1 个字符 | `re.match("[abc]", "b")` | 匹配成功（"b"） |
| `[0-9]` | 匹配任意 1 个数字（等价于 `\d`） | `re.match("[0-9]", "3")` | 匹配成功（"3"） |
| `[a-z]` | 匹配任意 1 个小写字母 | `re.match("[a-z]", "x")` | 匹配成功（"x"） |
| `[A-Z]` | 匹配任意 1 个大写字母 | `re.match("[A-Z]", "Y")` | 匹配成功（"Y"） |
| `[a-zA-Z]` | 匹配任意 1 个字母 | `re.match("[a-zA-Z]", "M")` | 匹配成功（"M"） |
| `[^0-9]` | 匹配任意 1 个非数字（`^` 在 `[]` 中表示“取反”） | `re.match("[^0-9]", "a")` | 匹配成功（"a"） |
| `\d` | 匹配任意 1 个数字（等价于 `[0-9]`） | `re.match(r"\d", "8")` | 匹配成功（"8"） |  
| `\D` | 匹配任意 1 个非数字（等价于 `[^0-9]`） | `re.match(r"\D", "!")` | 匹配成功（"!"） |
| `\s` | 匹配任意 1 个空白字符（空格、制表符、换行符） | `re.match(r"\s", " ")` | 匹配成功（空格） |
| `\S` | 匹配任意 1 个非空白字符 | `re.match(r"\S", "abc")` | 匹配成功（"a"） |
| `\w` | 匹配字母、数字、汉字、下划线（常用） | `re.match(r"\w", "哈")` | 匹配成功（"哈"） |
| `\W` | 匹配非单词字符（如 `@`、`#`、`!`） | `re.match(r"\W", "@")` | 匹配成功（"@"） |

### 核心规则：匹配多个字符

单个字符规则只能匹配 1 个字符，结合“数量限定符”，可匹配多个字符，覆盖复杂场景：

| 规则符号 | 通俗含义 | 示例 | 匹配结果 |
|----------|----------|------|----------|
| `*` | 匹配前一个字符 0 次或多次 | `re.match(r"\d*", "123abc")` | 匹配成功（"123"） |
| `+` | 匹配前一个字符 1 次或多次 | `re.match(r"\d+", "123abc")` | 匹配成功（"123"） |
| `?` | 匹配前一个字符 0 次或 1 次 | `re.match(r"\d?", "123")` | 匹配成功（"1"） |
| `{m}` | 匹配前一个字符恰好 m 次 | `re.match(r"\d{3}", "1234")` | 匹配成功（"123"） |
| `{m,n}` | 匹配前一个字符至少 m 次，最多 n 次（m ≤ n） | `re.match(r"\d{1,3}", "1234")` | 匹配成功（"123"） |

````python
res = re.match(r"^\w+@\w+\.[a-z]{2,4}$", "example@example.com")    # ^表示字符串开头，$表示字符串结尾，{2,4}表示匹配前一个字符2到4次
if res:
    print("Valid email address")
else:
    print("Invalid email address")
````

### 1.6  
核心规则：匹配分组（进阶）

用 `()` 将部分规则分组，可实现“多选一”或“引用分组结果”，适合复杂匹配场景。

### 1. 多选一（`|`）

**规则：**  
`(规则1|规则2|规则3)`，匹配任意一个符合的规则。  

### 2. 引用分组结果（`\num`）  

用 `\num` 引用前面第 `num` 个分组匹配的内容（比如匹配成对的 HTML 标签）。

````python
res = re.match(r"<(\w+)>.*</\1>", "<tag>content</tag>")
print(res.group())    # 输出tag
````

## 1.7 核心总结
1. **正则表达式**：描述字符串模式的规则，用于匹配、验证、提取字符串。

2. **核心流程**：导入 `re` 模块 → 定义规则（pattern）→ 执行匹配 → 提取结果。

3. **必记规则：**
   - **单个字符**：`.`（任意）、`\d`（数字）、`\w`（单词字符）、`[a-z]`（小写字母）；
   - **数量限定**：`*`（0+ 次）、`+`（1+ 次）、`?`（0-1 次）、`{m,n}`（m-n 次）；
   - **位置限定**：`^`（开头）、`$`（结尾）；
   - **分组**：`()`（分组）、`|`（多选一）、`\num`（引用分组）。

4. **关键提醒：**
   - 匹配失败返回 `None`，需先判断再调用 `group()`；
   - 带 `\` 的规则加 `r` 前缀避免转义。

# 2. 常用函数

````python
import re

# 开头匹配成功
res1 = re.match("th", "this is python")
print(res1.group())  # 输出: th

# 开头不匹配，返回 None
res2 = re.match("th", "python is good")
print(res2)  # 输出: None
````

## 2.1 核心函数 `re.search()` —— 全局找第一个匹配

### 2.1.1 作用

`search()` 会遍历整个字符串，不管开头是否匹配，只要找到第一个符合规则的子串就返回结果；  
如果遍历完整个字符串都没找到，则返回 `None`。

👉 类比理解：  
就像在一本书里找“苹果”这个词，从第一页读到最后一页，找到第一个“苹果”就停下。

---

### 2.1.2 基本语法

```python
re.search(pattern, string, flags=0)

参数说明
- pattern：正则匹配规则
- string：要搜索的原始字符串
- flags：可选参数（控制匹配方式，如忽略大小写等）

## 2.3 核心函数：`re.findall()` —— 找所有匹配项（返回列表）

### 2.3.1 作用

`findall()` 会遍历整个字符串，找到**所有符合规则的非重叠子串**，并以列表形式返回；  
如果没有匹配项，返回空列表 `[]`（不会返回 `None`）。

👉 类比理解：  
就像在一本书里找“苹果”，把所有出现“苹果”的地方全部标记出来，整理成清单。


### 2.3.2 基本语法

```python
re.findall(pattern, string, flags=0)


````python
res=re.findall(r"\d","321fsasad")
print(res)    # 输出: ['3', '2', '1']
````

### 2.3.3 应用场景

- 提取文本中所有符合规则的内容  
  （如：所有手机号、所有邮箱、所有日期）

- 批量收集关键信息  
  （如：从日志中提取所有错误码）

## 2.4 核心函数 `re.sub()` —— 替换匹配项

### 2.4.1 作用

`sub()` 是 "substitute"（替换）的缩写，会遍历字符串，  
把所有符合规则的子串替换成指定内容；可以控制替换次数，默认替换所有匹配项。

👉 类比理解：  
就像在文档中把所有“苹果”替换成“橙子”，  
也可以只替换前 2 个，而不是全部替换。

---

### 2.4.2 基本语法

```python
re.sub(pattern, repl, string, count=0, flags=0)

**参数说明**
- `pattern`：正则匹配规则
- `repl`：要替换成的新内容（字符串或函数）
- `string`：要处理的原始字符串
- `count`：可选参数  
    默认 `0` → 替换所有匹配项  
    设为 `1` → 只替换第一个
- `flags`：可选参数（控制匹配方式）

### 2.4.3 应用场景

- 过滤敏感字符  
  （如：把文本中的脏话替换成 `*`）

- 格式化文本  
  （如：把日期中的 `-` 替换成 `/`）

- 清理无效字符  
  （如：删除文本中的特殊符号、空白字符）

# 2.5 核心总结  
### 1. 三个核心函数的用途

- `re.search()`  
  全局找第一个匹配，返回 `Match` 对象（失败返回 `None`）

- `re.findall()`  
  全局找所有匹配，返回列表（失败返回空列表 `[]`）

- `re.sub()`  
  替换匹配项，返回替换后的新字符串（默认替换所有）


### 2. 关键区别

- `match()` vs `search()`  
  - `match()`：只匹配开头  
  - `search()`：全局查找第一个  

- `findall()`  
  - 返回列表  
  - 不需要判断是否为 `None`（空列表也不会报错）


### 3. 实用技巧

- 提取信息 → 用 `findall()`  
  （如：所有手机号、邮箱）

- 单个匹配 → 用 `search()`  
  （如：找第一个日期）

- 验证格式 → 用 `match()`  
  （如：手机号、邮箱格式验证）

- 替换 / 过滤 → 用 `sub()`  
  （如：打码、清理特殊字符）

# 3. 贪婪匹配 & 非贪婪匹配

## 3.1 什么是贪婪匹配 / 非贪婪匹配？

用一个生活例子帮你理解：

- **贪婪模式**：  
  给你一个袋子让你装苹果，能装多少装多少，直到装不下（尽可能多拿）

- **非贪婪模式**：  
  给你一个袋子，只要装够最低要求就行，多一个都不装（尽可能少拿）


### 对应正则规则

- **贪婪模式（默认）**  
  量词（如 `*`、`+`、`{m,n}`）会匹配尽可能多的字符，直到无法继续匹配

- **非贪婪模式**  
  在量词后加 `?`，表示匹配尽可能少的字符，只要满足规则即可

### 对比示例

````python
text = "<div>hello</div><div>world</div>"

# 贪婪匹配
res1 = re.findall(r"<div>.*</div>", text)
print(res1)
# 输出: ['<div>hello</div><div>world</div>']

# 非贪婪匹配
res2 = re.findall(r"<div>.*?</div>", text)
print(res2)
# 输出: ['<div>hello</div>', '<div>world</div>']
````

## 3.2 关键对比：同一规则的两种模式结果

用表格清晰展示，一目了然：

| 正则规则（贪婪） | 正则规则（非贪婪） | 匹配字符串     | 贪婪模式结果       | 非贪婪模式结果     |
|------------------|------------------|----------------|--------------------|--------------------|
| `\d*`           | `\d*?`          | `123456abc`    | `123456`（拿完所有数字） | `""`（拿 0 个）     |
| `\d+`           | `\d+?`          | `123456abc`    | `123456`（拿完所有数字） | `1`（拿 1 个）      |
| `\d{2,5}`       | `\d{2,5}?`      | `123456abc`    | `12345`（拿满 5 个）     | `12`（拿 2 个）     |
| `.*`            | `.*?`           | `a1b2c3`       | `a1b2c3`（拿完所有字符） | `""`（拿 0 个）     |
| `a.*b`          | `a.*?b`         | `a1b2c3b`      | `a1b2c3b`（从第一个 a 到最后一个 b） | `a1b`（从第一个 a 到最近的 b） |

### 🔑 核心理解
- **贪婪模式（默认）**：尽可能多匹配  
- **非贪婪模式（加 `?`）**：尽可能少匹配（满足条件即可）

## 3.3 核心总结

### 1. 模式定义
- **贪婪模式（默认）**：量词尽可能多匹配，无需额外符号；
- **非贪婪模式**：在量词后加 `?`，尽可能少匹配。

### 2. 影响范围
- 只对量词生效：`*`、`+`、`{m,n}`、`?`

### 3. 核心区别
- **贪婪**：能多拿就多拿，直到无法继续；
- **非贪婪**：能少拿就少拿，满足最低要求即可。

### 4. 实用技巧
- 提取“最近匹配”“单个片段” → 用非贪婪（加 `?`）；
- 提取“完整片段”“所有符合内容” → 用贪婪（默认）；
- 不确定时：两种模式都试一下，对比结果即可。
