# 1. 文件操作（持久化存储）  
## 1.1 文件的两种类型  
| 类型    | 通俗理解       | 存储内容           | 查看方式               | 常见扩展名                       |
| ----- | ---------- | -------------- | ------------------ | --------------------------- |
| 文本文件  | 人类可读的文件    | 字符（按编码转为二进制存储） | 记事本、VS Code 等文本编辑器 | `.txt`、`.py`、`.md`          |
| 二进制文件 | 计算机直接处理的文件 | 原始二进制数据（非字符）   | 专用软件（图片查看器、播放器等）   | `.jpg`、`.png`、`.mp3`、`.exe` |


## 1.2 文件的操作三步骤  
无论读还是写文件，都遵循 “打开 → 操作 → 关闭” 的流程：

1. **打开文件**：用 `open()` 创建文件对象，建立程序与文件的连接。  
2. **读写操作**：  
   - 读（将文件内容加载到内存）  
   - 写（将内存数据存入文件）  
3. **关闭文件**：用 `close()` 释放系统资源，避免占用内存或文件冲突。

---


## 2. 核心操作 1：打开与关闭文件

### 2.1 打开文件：`open()`

**基本语法：**
```python
文件对象 = open(文件路径, 访问模式, encoding="编码格式")



3 个关键参数：  

文件路径：文件在电脑上的位置（如 C:\test.txt）；  
访问模式：控制文件读写权限（如 r 只读、w 只写，后面详细讲）；  
encoding：文本文件的编码格式（推荐 utf-8，避免中文乱码，二进制文件不用加）。  

### 2.2 关闭文件：`close()`  

### 2.3 更安全的方式：`with` 语句（推荐）

用 `with` 语句打开文件，代码块执行完后会自动关闭文件，不用手动写 `close()`，避免遗漏。

---


## 3. 核心操作 2：文件读写（重点）

### 3.1 文件访问模式

不同模式控制读写权限，避免报错：

| 模式 | 功能描述 | 是否创建新文件 | 注意事项 |
|------|----------|----------------|----------|
| `r`  | 只读（默认） | 否（文件不存在报错） | 只能读，不能写 |
| `w`  | 只写 | 是（文件不存在创建，存在则覆盖内容） | 只能写，不能读 |
| `a`  | 追加 | 是（文件不存在创建，存在则写在末尾） | 只能写，不能读 |
| `r+` | 读写 | 否（文件不存在报错） | 可同时读写，不覆盖原有内容 |
| `w+` | 读写 | 是（文件不存在创建，存在则覆盖） | 可同时读写 |
| `a+` | 读写 | 是（文件不存在创建，存在则追加） | 可同时读写，写在末尾 |  

**注意**: 二进制文件需加b模式（`rb`、`wb`等）


### 3.2 读取文件：3 种常用方法

适用于文本文件，二进制文件读取用 `read()` 即可。

#### 3.2.1 `read(n)`：读取指定字符 / 字节

**语法：**
```python
文件对象.read(n)；  
省略n则读取整个内容。   


#### 3.2.2 `readline()` : 逐行读取  
每次读取1行，文件指针自动移到下一行，适合大文件(避免占用过多内存)；  
读取到文件末尾返回空字符。



#### 3.2.3 `readlines()` : 读取所有行（返回列表）  
把文件所有行作为列表元素返回，每行是一个字符串，适合小文件。  


### 3.3 写入文件：`write()`  
- 文本文件:内容必须是字符串;  
- 进制文件:内容必须是字节串(bytes类型)。


### 3.4 进阶：文件指针定位（了解）

文件指针是“下次读写的起始位置”，默认：

- **读模式**：指针在文件开头；
- **写 / 追加模式**：指针在文件末尾。

**常用方法：**

1. `tell()`：返回当前指针位置（字节数）；
2. `seek(offset, whence)`：移动指针  
   - `offset`：偏移量  
   - `whence`：参考点  
     - `0`：文件开头  
     - `1`：当前位置  
     - `2`：文件末尾  

## 4. 避坑指南：中文乱码与常见错误

### 4.1 中文乱码：必加 `encoding='utf-8'`  
Windows系统默认编码是`GBK`，Python读取中文文件时，必须指定`encoding="utf-8"`，否则会乱码或报错。  

### 4.2 常见错误及解决

1. **FileNotFoundError**：文件不存在  
   （检查路径是否正确，或用 `w` / `a` 模式创建）

2. **io.UnsupportedOperation**：读写权限不匹配  
   （比如 `r` 模式下尝试写，改模式为 `r+`）

3. **UnicodeDecodeError**：编码不匹配  
   （指定正确的 `encoding`，如 `utf-8`）  


## 5. 核心总结

1. **文件操作流程：**  
   打开（`open()` / `with`） → 读写（`read()` / `write()`） → 关闭（自动 / `close()`）

2. **关键参数：**  
   - **访问模式**：`r` 只读、`w` 只写（覆盖）、`a` 追加、`rb` / `wb` 二进制读写；  
   - **encoding**：文本文件用 `utf-8`，二进制文件不用加。

3. **读取方法：**  
   - `read(n)`：读指定字符 / 字节，适合小文件；  
   - `readline()`：逐行读，适合大文件；  
   - `readlines()`：读所有行（列表），适合小文件。

4. **写入方法：**  
   - `write()`：文本写字符串，二进制写字节串。

5. **推荐用法：**  
   使用 `with` 语句自动关闭文件，避免资源泄露。

# 2. 文件路径


## 2.1 绝对路径  
绝对路径是从电脑“根目录”(比如Windows的C盘、D盘，Mac的/Users)开始，到文件的完整地址，包含所有文件夹层级，不依赖任何“参考位置”。

**提示**: Windows系统中，路径用反斜杠`\`，但Python中`\`是转义符，所以要么加`r`前缀(推荐)，要么用双反斜杠`\\`。

**示例：**

```python
# 正确写法1：加 r 前缀（原始字符串，不转义）
f = open(r"D:\PythonStudy\test.txt", "r", encoding="utf-8")

# 正确写法2：双反斜杠
f = open("D:\\PythonStudy\\test.txt", "r", encoding="utf-8")

### 2.1.1 优点 & 缺点

| 优点 | 缺点 |
|------|------|
| **位置唯一**：<br>无论程序在哪里运行，都能准确找到文件 | 路径太长，写起来麻烦 |
| 不受当前目录影响，稳定性高 | 移植性差：换电脑 / 改文件夹结构，路径就失效 |

### 2.1.2 应用场景
- 访问系统固定位置的文件(比如C盘的配置文件、D盘的日志文件);  
- 需要确保在任何环境下都能找到文件的场景(比如脚本单独运行，不依赖项目目录)

## 2.3 相对路径: 文件的"相对位置"  
相对路径是相对于"当前工作目录"(Python脚本运行的目录)的地址，不用写完整根目录，只写"从当前目录出发的路径"。

### 2.3.1 核心符号

相对路径用两个符号表示“层级关系”：

- `.` ：代表“当前目录”（脚本所在的文件夹），可省略；
- `..`：代表“上一级目录”（脚本所在文件夹的父文件夹）。

### 2.3.2 优点 & 缺点

| 优点 | 缺点 |
|------|------|
| 路径简洁，写起来方便 | 依赖当前目录：换了脚本运行位置，路径就失效 |
| 移植性强：项目文件夹整体移动，路径依然有效 | 不直观：需要清楚当前目录和文件的层级关系 |

### 2.3.3 应用场景

- 项目内部文件操作（比如脚本和数据文件在同一个项目文件夹里）；
- 团队协作、代码部署：项目文件夹整体迁移时，不用修改路径。

# 3.目录常用操作  
*引言*: 之前学了文件的读写，现在来掌握文件和目录的“管理操作”：  
比如重命名文件、删除文件创建文件夹、查看目录内容等。  
这些操作不用手动打开文件夹逐个点击，Python的`os`模块就能帮们一键搞定。  
借助它提供的方法，程序可以自动完成这些管理工作，既能避免手动操作的繁琐，还能误操作，新手也能轻松实现文件目录的自动化管理。

## 3.1 核心：`os`模块

````python
import os       # python自带
````

## 3.2. 文件操作：重命名与删除

### 3.2.1 重命名文件 / 目录：`os.rename()`

#### 1 核心功能
给文件或目录改名字，还能实现“移动文件到同一磁盘的其他目录”
（本质是改路径 + 改名字）。

#### 2 基本语法
```python
os.rename(源路径, 目标路径)  

参数详解：

- 源路径：文件 / 目录的原始路径（相对路径或绝对路径）；
- 目标路径：修改后的新路径（如果只改名字，路径不变；如果移动，改路径即可）。

```python
import os

# 示例1：重命名当前目录的文件（只改名字，不改路径）
# 把 "image[备份].png" 改成 "image2.png"
os.rename("image[备份].png", "image2.png")

# 示例2：移动文件到同一磁盘的其他目录（改路径，名字不变）
# 把当前目录的 "image2.png" 移动到 D 盘 tools 目录
os.rename("image2.png", r"D:\tools\image2.png")

# 示例3：重命名目录（和文件操作一致）
# 把当前目录的 "old_dir" 改成 "new_dir"
os.rename("old_dir", "new_dir")

### 避坑注意

- 源路径必须存在，否则报错 `FileNotFoundError`（文件 / 目录不存在）；
- 不支持跨磁盘移动（比如从 D 盘移到 E 盘），会报错 `WinError`；
- 目标路径如果已存在同名文件 / 目录，会直接覆盖（谨慎操作！）。

## 3.2.2 删除文件：`os.remove`

### 1. 核心功能  
删除指定路径的文件（不可删目录）  

### 2. 基本语法   
`os.remove(文件路径)`

## 3.3 目录操作：创建删除与查看  
### 3.3.1 创建目录：`os.mkdir(目录路径)`  
### 3.3.2 删除目录：`os.rmdir(目录路径)`  
### 3.3.2 查看当前工作目录：`os.getcwd()`  
### 3.3.2 查看目录内容：`os.listdir(目录路径)`

# 4.路径检验:判断文件/目录是否存在
## 4.1 检验是否存在：`os.path.exists()`
### 4.1.1 核心功能  
判断指定路径(文件或目录)是否存在，返回`True or False`  
## 4.2 检验是否为文件：`os.path.isfile()`  
### 4.2.1 核心功能  
判断指定路径是否存在，且是一个文件，返回`True or False`  
## 4.3 检验是否为目录：`os.path.exists()`  

## 4.4 核心总结
1. 所有操作需先导入 `os` 模块（`import os`）。

2. **文件操作：**
   - 重命名 / 移动：`os.rename(源路径, 目标路径)`（同磁盘可用）；
   - 删除文件：`os.remove(文件路径)`（只能删文件，谨慎！）。

3. **目录操作：**
   - 创建目录：`os.mkdir(目录路径)`（不能已存在）；
   - 删除目录：`os.rmdir(目录路径)`（只能删空目录，谨慎！）；
   - 查看当前目录：`os.getcwd()`（返回绝对路径）；
   - 查看目录内容：`os.listdir(目录路径)`（返回文件 / 目录列表）。

4. **路径检验（避免报错）：**
   - 路径是否存在：`os.path.exists(路径)`；
   - 是否为文件：`os.path.isfile(路径)`；
   - 是否为目录：`os.path.isdir(路径)`。

5. **关键提醒：**
   - 删除操作不可逆，建议先判断路径存在性再操作。
